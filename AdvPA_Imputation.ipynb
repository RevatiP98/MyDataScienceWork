#Chemical Process Dataset

```python
chem_df=pd.read_csv("/content/ChemicalProcess.csv")
chem_df
```

```python
chem_df.info()
```

```python
chem_df.isnull().sum()
```

```python
chem_df.columns[chem_df.isnull().any()]
```

```python
from statistics import mean, mode, median
```

```python
imputer1=SimpleImputer(strategy='mean')
```

```python
imputer1.fit(chem_df)
```

```python
imp_chem=imputer1.transform(chem_df)
imp_chem
```

```python
pd_chem=pd.DataFrame(imp_chem, columns=chem_df.columns)
pd_chem
```

```python
pd_chem.columns[pd_chem.isnull().any()]
```

```python
pd_chem.columns
```

```python
X=pd_chem.drop('Yield',axis=1)
```

```python
Y=pd_chem['Yield']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,Y, test_size=0.3, random_state=2022)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2_Score: with 'Mean' as a Strategy :",r2_score(y_test,y_pred))
```
